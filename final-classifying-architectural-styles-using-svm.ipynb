```python
# Extracting HOG and LBP features using skimage

from skimage.feature import hog, local_binary_pattern

def extract_HOG_features(img):
    hog_image = hog(img)
    return hog_image

def extract_LBP_features(img):
    lbp = local_binary_pattern(img, 8, 1)
    return lbp
```

```python
images_grayscale[0].astype('uint8')
```

```python
# Extraction of SIFT features

def extract_SIFT_features(img):
    sift = cv2.SIFT_create()
    kp, des = sift.detectAndCompute(img, None)
    return des

def bag_of_features(features, centres, k = 500):
    vec = np.zeros((1, k))
    for i in range(features.shape[0]):
        feat = features[i]
        diff = np.tile(feat, (k, 1)) - centres
        dist = pow(((pow(diff, 2)).sum(axis = 1)), 0.5)
        idx_dist = dist.argsort()
        idx = idx_dist[0]
        vec[0][idx] += 1
    return vec

SIFT_features = list()
for i in np.array(images_grayscale).astype('uint8'):
    SIFT_features.append(extract_SIFT_features(i))
SIFT_features = np.vstack(SIFT_features)

k = 150
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 0.1)
flags = cv2.KMEANS_RANDOM_CENTERS
compactness, lbl, centres = cv2.kmeans(SIFT_features, k, None, criteria, 10, flags)

image_descriptors = list()
images_SIFT_features = list()

for i in np.array(images_grayscale).astype('uint8'):
    descriptor = extract_SIFT_features(i)
    if descriptor is not None:
        img_vector = bag_of_features(descriptor, centres, k)
        images_SIFT_features.append(img_vector)
```

```python
from sklearn.model_selection import train_test_split

train_images_grayscale, test_images_grayscale, train_labels_grayscale, test_labels_grayscale = train_test_split( # grayscale features
                                                                    input_images_grayscale, output_labels,
                                                                    random_state = 42,
                                                                    test_size = 0.2)

train_images_hog, test_images_hog, train_labels_hog, test_labels_hog = train_test_split( # HOG features
                                                                    input_images_hog, output_labels,
                                                                    random_state = 42,
                                                                    test_size = 0.2)

train_images_lbp, test_images_lbp, train_labels_lbp, test_labels_lbp = train_test_split( # LBP features
                                                                    input_images_lbp, output_labels,
                                                                    random_state = 42,
                                                                    test_size = 0.2)

train_images_sift, test_images_sift, train_labels_sift, test_labels_sift = train_test_split( # SIFT features
                                                                    input_images_sift, output_labels,
                                                                    random_state = 42,
                                                                    test_size = 0.2)

train_images, test_images, train_labels, test_labels = train_test_split( # RGB features
                                                                    input_images, output_labels,
                                                                    random_state = 42,
                                                                    test_size = 0.2)
```
